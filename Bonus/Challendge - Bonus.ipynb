image.png

```python
AQI = epa2015['AQI']
bins = [0, 50, 100, 150, 200, 300, float('inf')]
bin_names = ['Good', 'Moderate', 'Unhealthy-sensitive', 'Unhealthy', 'Vert Unhealthy', 'Hazardous']
epa2015['AQI category'] = pd.cut(AQI, bins, labels = bin_names)
epa2015.head()
```

```python
AQI2016 = epa2016['AQI']

epa2016['AQI category'] = pd.cut(AQI2016, bins, labels = bin_names)
epa2016.head()
```
