# Q1: Enter a Number

```python
n = input("Enter a Number:")
z = 0
odd = 0
even = 0
a = [int(d) for d in str(n)]
print(a)

for i in a:
  if int(i) == 0:
    z += 1
  elif int(i)%2 == 0:
    even += 1
  else:
    odd += 1
print("Odd Digits", odd)
print("Even Digits",even)
print("Zero Digits",z)
```

# Q2: Hailstone Sequence

```python
n = eval(input("Enter a Number"))
while True:
  if n % 2 ==0:
    n = n/2
    print(n)
  elif n % 2 ==1:
    n = (3*n) + 1
    print(n)
```

# Q3 Necklace

```python
def generate_necklace(seed1, seed2):
    sequence = [seed1, seed2]
    while True:
        next_num = (sequence[-1] + sequence[-2]) % 10
        sequence.append(next_num)
        if sequence[-2:] == [seed1, seed2]:
            break
        if sequence[-1] == seed1 or sequence[-1] == seed2:
            break
    return sequence

seed1 = int(input("Enter the first seed number: "))
seed2 = int(input("Enter the second seed number: "))

necklace_sequence = generate_necklace(seed1, seed2)
print(necklace_sequence)




```

# Q6 Rock Paper Scissor game


```python
from IPython.lib.pretty import Breakable
import random
game = ['A', 'B', 'C']
print("A:", "Paper")
print("B:", "Scissor")
print("C:", "Stone")
while True:
  a = input("please Enter A, B, C: ")
  b = random.choice(game)
  print("Computer", b)
  if a == b:
    print("same Result:", a, b)
    continue
  elif a == 'B' and b == "A":
    print("you win")
    break
  elif a =='A' and b =="C":
    print("you win")
    break
  elif a == "C" and b =="B":
    print('you win')
    break
  else:
    print('you lost')
    break
```
