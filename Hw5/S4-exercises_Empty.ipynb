## Exercise-3

```python
world = pd.read_csv("/content/drive/MyDrive/My Pitt/Python Programming/5mins_Project/worlddata.csv")
```

```python
import folium
map1 = folium.Map(location=[48, -102], zoom_start=3)
```

```python
url = 'https://raw.githubusercontent.com/python-visualization/folium/master/examples/data'
state_geo = f'{url}/us-states.json' # for US map with state boundaries
world_geo = f'{url}/world-countries.json' # for world map with country boundaries
state_unemployment = f'{url}/US_Unemployment_Oct2012.csv'
```

```python
Ageselect = world[(world['IndicatorName']=="Age dependency ratio (% of working-age population)") \
                       & (world['Year']==1960)]
Ageselect.head()
```

```python
data_to_plot = Ageselect[['CountryCode','Value']]
data_to_plot.head()
```

```python
map1 = folium.Map(location=[40, 0], zoom_start=0.5)
```

```python
folium.Choropleth(geo_data=world_geo, data=data_to_plot,
             name='choropleth',
             columns=['CountryCode', 'Value'],
             key_on='feature.id',
             fill_color='YlGnBu', fill_opacity=0.7, line_opacity=0.2,
             legend_name='Age dependency ratio (% of working-age population)').add_to(map1)
folium.LayerControl().add_to(map1)
map1
```
